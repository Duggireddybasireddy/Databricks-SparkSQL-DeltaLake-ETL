### What is the total number of trips recorded for each vendor ID, ordered from most trips to fewest

```sql
select vendorID, count(*) as `Number of trips` from yellow_filtered_delta
group by 1
order by 2 desc;
```

### Find total distance travelled by each vendor.

```sql
select VendorID, round(sum(trip_distance), 2) as `Total Distance Travelled` from yellow_filtered_delta
group by 1
order by 2 desc;
```

### what is the fare_amount of each vendor for each location.

```sql
select VendorID, pickuplocationID, round(sum(fare_amount), 2) as `Total_Fare` from yellow_filtered_delta
group by 1, 2
order by 3 desc;
```

```sql
create table vendor_travel_time as 
select VendorID, dropLocationID, 
  sum(timestampdiff(HOUR, pickup_datetime, dropoff_datetime)) as `Total_Time_Hrs` 
  from yellow_filtered_delta
  group by  VendorID, droplocationID
  order by VendorID;
```

# **Report on Combined Delta Table**

### **1.Generate Report to get total trip time in hours for each taxi type**


```sql
select Taxitype, VendorID, count(*) as `Total Trips` from yellowgreentripcombinedelta
group by 1, 2
order by 3;
```

### 2.	Generate Report to get taxi-type-wise total trip time in hours for each trip-month 

```sql
select Taxitype, month(pickup_datetime) as `Trip Month`, 
sum(timestampdiff(HOUR, pickup_datetime, dropoff_datetime)) 
from yellowgreentripcombinedelta 
group by 1, 2 
order by 2;
```

### 3.	Generate Report to get taxi-type-wise total number of passengers for each trip month


```sql
select Taxitype, date_format(pickup_datetime, 'MMMM') as `Trip Month`, sum(passenger_count) as `Total Passengers` 
from yellowgreentripcombinedelta 
group by 1, 2
order by 3;
```

### 4.	Generate Report to get taxi-type-wise total number of payments for each payment-type


```sql

select Taxitype, payment_type, count(*) as `Total Payments` from yellowgreentripcombinedelta
group by 1, 2
order by 3;
```

### 5.	Generate Report to get  the total number of light trips for each taxi type (when passenger count for each trip is <=2)


```sql

select Taxitype, count(*) as `Light Trips` from yellowgreentripcombinedelta
where passenger_count <= 2
group by 1
order by 2;
```

### 6.	Generate Report to get the total number of light trips for each taxi type month-wise (when passenger count for each trip is <=2) and save the results in a delta table lightTripsTaxiTypeMonWise


```sql

create or replace table lightTripsTaxiTypeMonWise as
(select Taxitype, date_format(pickup_datetime, 'MMM') as `Trip_Month`, count(*) as `Light_Trips` 
from yellowgreentripcombinedelta
where passenger_count <= 2
group by 1, 2
order by 2); 
```

```sql
select * from lightTripsTaxiTypeMonWise;
```

### 7.	Generate Report to get  the total number of fully-loaded trips for each taxi type (when passenger count for each trip is >=4)


```sql
select Taxitype, count(*) as `Full Loaded Trips` from yellowgreentripcombinedelta
where passenger_count >= 4
group by 1
order by 2;
```

### 8.	Generate Report to get  the total number of fully-loaded trips for each taxi type month-wise (when passenger count for each trip is >=4) and save the results in a delta table loadedTripsTaxiTypeMonWise


```sql
create or replace table loadedTripsTaxiTypeMonWise AS
(select Taxitype, date_format(pickup_datetime, "MMMM") as `Trip_Month`,count(*) as `Full_Loaded_Trips` 
from yellowgreentripcombinedelta
where passenger_count >= 4
group by 1, 2
order by 3)
```

```sql
select * from loadedTripsTaxiTypeMonWise;
```

### 9.	Generate Report to get  the total number of midnight trips for each taxi type (when trips happen between 12AM to 4AM)


```sql
select TaxiType, date_format(pickup_datetime, "HH") as `Trip Time`, count(*) as `Total Trips` 
from yellowgreentripcombinedelta
where date_format(pickup_datetime, "HH") between 0 and 4
group by 1, 2
order by 3;
```

### 10.	Generate Report to get the total number of midnight trips for each taxi type month-wise(when trips happen between 12AM to 4AM)


```sql
select TaxiType, date_format(pickup_datetime, "MMMM") as `Trip Month`, count(*) as `Total Trips`
from yellowgreentripcombinedelta
where date_format(pickup_datetime, "HH") Between 0 and 4
group by 1, 2
order by 3;
```
